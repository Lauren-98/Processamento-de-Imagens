```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# implementando um aramzenamento de número de pixels para cada nível de intensidade
hist_data = {}

# input com o número de pixels de cada nível
for i in range(10):
    nivel = int(input(f"Insira o número de pixels para o nível {i}: "))
    hist_data[i] = nivel

# vetor de intensidades de cinza com base nos dados inseridos
pixels = []
for nivel, count in hist_data.items():
    pixels.extend([nivel] * count)

# calculando o histograma
hist, bins = np.histogram(pixels, bins=10, range=(0, 10))

# plotando o histograma
plt.bar(bins[:-1], hist, width=0.8)
plt.title("Histograma de 10 Níveis")
plt.xlabel("Níveis de Intensidade de Cinza")
plt.ylabel("Frequência")
plt.show()
```

```python
# fazendo sem o input

freq = [150, 250, 200, 300, 250, 250, 1000, 1000, 2500, 4100]

plt.bar(range(10), freq, width=0.8, color='g', alpha=0.6)
plt.title("Histograma Equalizado")
plt.xlabel("Níveis de Intensidade de Cinza")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

```python
# calculando a equalização

L = 10

n_l = np.array([150, 250, 200, 300, 250, 250, 1000, 1000, 2500, 4100])

results = []

for r in range(10):
    sum_n_l = np.sum(n_l[:r+1])

    S = ((L - 1) / (n_l.sum())) * sum_n_l
    results.append(S)

    print(f"Para r = {r}, S = {S}")
```

```python
# calculando a equalização com valores arredondados pelo "round"

L = 10

n_l = np.array([150, 250, 200, 300, 250, 250, 1000, 1000, 2500, 4100])

results = []

for r in range(10):
    sum_n_l = np.sum(n_l[:r+1])

    S = round(((L - 1) / (n_l.sum())) * sum_n_l)
    results.append(S)

    print(f"Para r = {r}, S = {S}")
```

```python
# calculando a equalização e plotando os resultados

n_l = np.array([150, 250, 200, 300, 250, 250, 1000, 1000, 2500, 4100])

L = 10

results = []

# histograma equalizado
hist_equalizado = np.zeros(L, dtype=int)

for r in range(10):
    sum_n_l = np.sum(n_l[:r+1])

    S = round(((L - 1) / (n_l.sum())) * sum_n_l)
    results.append(S)

    hist_equalizado[S] += n_l[r]

    print(f"Para r = {r}, S = {S}")


x = list(range(L))

# histograma original
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(x, n_l, width=0.8, color='black', alpha=0.7)
plt.title('Histograma Original')
plt.xlabel('Nível de Cinza')
plt.ylabel('Frequência')
plt.grid(True)

# histograma equalizado
plt.subplot(1, 2, 2)
plt.bar(x, hist_equalizado, width=0.8, color='g', alpha=0.7)
plt.title('Histograma Equalizado')
plt.xlabel('Nível de Cinza')
plt.ylabel('Frequência')
plt.grid(True)

plt.tight_layout()
plt.show()
```
